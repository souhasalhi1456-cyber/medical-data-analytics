```python
def nettoyer_colonnes(df):
    for col_index in range(len(df.columns) - 2, len(df.columns)): 
        total = 0
        count = 0
        col = df.columns[col_index]
        for i in range(len(df[col])):
            value = df.iloc[i, col_index]
            if not pd.isnull(value):  # Vérify if valye is nan
                total += value
                count += 1
        if count != 0:
            moyenne_colonne = total / count
            for i in range(len(df[col])):
                value = df.iloc[i, col_index]
                if pd.isnull(value):
                    df.iloc[i, col_index] = moyenne_colonne
    return df
df2018=nettoyer_colonnes(df2018)
df2019=nettoyer_colonnes(df2019)
```

```python
def renommer_colonnes(df, annee,c1,c2):
    nouveaux_noms = {c1: 'CO2_' + str(annee), c2: 'MF_' + str(annee)}
    df.rename(columns=nouveaux_noms, inplace=True)
    return df
df2018 = renommer_colonnes(df2018, 2018,"CO2 emission 2018","Material footprint 2018")
df2019 = renommer_colonnes(df2019, 2019,"CO2 emission 2019","material foot per cap(tonnes) 2019")
```

```python
def calculate_IE(MAE, AEA):
    return (MAE / 15 + AEA / 18) / 2

def calculate_IEV(EV):
    return (EV - 20) / (85 - 20)
def calculate_IR(RNBph):
    return (math.log(RNBph) - math.log(100)) / (math.log(20000) - math.log(100))

def calculate_ID(IEV, IE, IR):
    return 3 * (IEV * IE * IR) ** (1/3)

def calculate_IIE(EM, CO2):
    DM = math.sqrt(max((EM / 6.52), 1) * max((CO2 / 1.58), 1))
    if DM <= 4:
        return 1 + (math.exp(DM) - math.exp(1)) / (math.exp(4) - math.exp(1))
    else:
        return DM - 2

def calculate_SDI(ID, IIE):
    return ID / IIE

```

```python
df2018['IEV_2018'] = df2018['Life expectency 2018'].apply(calculate_IEV)
df2018['IE_2018'] = df2018.apply(lambda row: calculate_IE(row['Mean years of scooling 2018'], row['Expect years of scooling 2018']), axis=1)
df2018['IR_2018'] = df2018['Gross income Level per capita 2018'].apply(calculate_IR)
df2018['ID_2018'] = df2018.apply(lambda row: calculate_ID(row['IEV_2018'], row['IE_2018'], row['IR_2018']), axis=1)
df2018['IIE_2018'] = df2018.apply(lambda row: calculate_IIE(row['MF_2018'], row['CO2_2018']), axis=1)
df2018['SDI_2018'] = calculate_SDI(df2018['ID_2018'], df2018['IIE_2018'])

df2019['IEV_2019'] = df2019['life expectency 2019'].apply(calculate_IEV)
df2019['IE_2019'] = df2019.apply(lambda row: calculate_IE(row['means years of schooling 2019'], row['expected years of scooling 2019']), axis=1)
df2019['IR_2019'] = df2019['GNI per capita 2019'].apply(calculate_IR)
df2019['ID_2019'] = df2019.apply(lambda row: calculate_ID(row['IEV_2019'], row['IE_2019'], row['IR_2019']), axis=1)
df2019['IIE_2019'] = df2019.apply(lambda row: calculate_IIE(row['MF_2019'], row['CO2_2019']), axis=1)
df2019['SDI_2019'] = calculate_SDI(df2019['ID_2019'], df2019['IIE_2019'])
```

```python
range_2019 = df2019['SDI_2019'].max() - df2019['SDI_2019'].min()

class_interval = range_2019 / 3

C1 = [df2019['SDI_2019'].min(), df2019['SDI_2019'].min() + class_interval]
C2 = [df2019['SDI_2019'].min() + class_interval, df2019['SDI_2019'].min() + 2 * class_interval]
C3 = [df2019['SDI_2019'].min() + 2 * class_interval, df2019['SDI_2019'].max()]  # Adjusted to ensure all values are included

def classify_SDI_level(SDI):
    if SDI >= C1[0] and SDI < C1[1]:
        return 'faible'
    elif SDI >= C2[0] and SDI < C2[1]:
        return 'moyen'
    elif SDI >= C3[0] and SDI <= C3[1]:  # Adjusted to include the upper bound
        return 'fort'

df_new['SDIL_2019'] = df_new['SDI_2019'].apply(classify_SDI_level)
```

```python
range_2019 = df_new['ID_2019'].max() - df2019['ID_2019'].min()

class_interval = range_2019 / 3

ID1 = [df_new['ID_2019'].min(), df2019['ID_2019'].min() + class_interval]
ID2 = [df_new['ID_2019'].min() + class_interval, df2019['ID_2019'].min() + 2 * class_interval]
ID3 = [df_new['ID_2019'].min() + 2 * class_interval, df2019['ID_2019'].max()]  # Adjusted to ensure all values are included

def classify_ID_level(ID):
    if ID >= ID1[0] and ID < ID1[1]:
        return 'faible'
    elif ID >= ID2[0] and ID < ID2[1]:
        return 'moyen'
    elif ID >= ID3[0] and ID <= ID3[1]:  # Adjusted to include the upper bound
        return 'fort'

df_new['DL_2019'] = df_new['ID_2019'].apply(classify_ID_level)
```

```python
# Ensure pandas is imported
import pandas as pd

# Sample DataFrame (replace this with your actual DataFrame loading/creation code)
# Example:
# df_new = pd.read_csv('your_data.csv')

# Step 1: Group the DataFrame by 'SDIL_2019' and 'DL_2019'
grouped_SDIL = df_new.groupby('SDIL_2019')
grouped_DL = df_new.groupby('DL_2019')

# Step 2: Retrieve the list of countries for each category
countries_SDIL = {group: set(group_df.index) for group, group_df in grouped_SDIL}
countries_DL = {group: set(group_df.index) for group, group_df in grouped_DL}

# Step 3: Initialize the correlation dictionary
correlation = {}

# Step 4: Compare pairwise the lists of countries for each category
for category_SDIL, countries_SDIL_list in countries_SDIL.items():
    for category_DL, countries_DL_list in countries_DL.items():
        # Compare categories and calculate common countries
        common_countries = countries_SDIL_list.intersection(countries_DL_list)
        # Store the result in the correlation dictionary
        correlation[(category_SDIL, category_DL)] = len(common_countries)

# Print the correlation dictionary
print("Correlation entre les categories SDIL-2019 et DL-2019 :")
for (category_SDIL, category_DL), common_countries_count in correlation.items():
    print(f"{category_SDIL} vs {category_DL}: {common_countries_count} common countries")
```

```python
def determine_opportunity(row):
    if row['DL_2019'] == 'fort' and row['SDIL_2019'] in ['moyen', 'faible']:
        return "à potentiel d’amélioration durable"
    elif row['SDIL_2019'] in ['fort', 'moyen'] and row['DL_2019'] == 'faible':
        return "à potentiel d’amélioration socio-économique"
    elif row['SDIL_2019'] in ['moyen', 'faible'] and row['DL_2019'] in ['faible', 'moyen']:
        return "à potentiel d’amélioration en developpement durable"
    else:
        return "pays à la pointe du développement durable"
df_new['Opportunité d’amélioration socio-économique et durable'] = df_new.apply(determine_opportunity, axis=1)
```
